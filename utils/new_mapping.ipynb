requests api

```python
def query(lat, lon):
    dataset = "data-es"
    base_url = f"https://equipements.sports.gouv.fr/api/explore/v2.1/catalog/datasets/{dataset}/records"

    params = {
        "select": "numinstallation,nominstallation,adresse,codepostal,new_name,new_code,nomequipement,coordgpsx,coordgpsy, carac72, typequipement",
        "where": f"within_distance(coordonnees, GEOM'POINT({lon} {lat})', 1000m)",
    }
    res = requests.get(url=base_url, params=params, timeout=2)
    print(res.status_code)
    df_res = pd.DataFrame(res.json()['results'])
    return df_res
```

```python
session_data.drop_duplicates(subset='lieu_epreuve', keep='first').reset_index(drop=True)
```

```python
list_df = []
for index, row in tqdm(session_data.drop_duplicates(subset='lieu_epreuve', keep='first').reset_index(drop=True).iterrows()):
    lat = row.latitude
    lon = row.longitude
    # print('longitude : ', lon)
    # print('latitude : ', lat)
    df_res = query(lat, lon)
    if df_res.any().any():
        df_res['infra1'] = row.NOM
        df_res = df_res.loc[df_res.carac72 != 0]
        # print('df_res : ', df_res)
        list_df.append(df_res)
```
